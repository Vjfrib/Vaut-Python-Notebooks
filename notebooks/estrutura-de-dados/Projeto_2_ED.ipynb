```python
import pdfplumber
from transformers import BertTokenizer, BertModel
import torch

# Função para extrair texto de um PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Carregar o tokenizer e o modelo BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Função para processar texto com BERT
def process_text_with_bert(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs)
    return outputs

# Substitua o caminho do arquivo PDF pelo caminho real
pdf_path = "Art2ao11.pdf"

# Extrair texto do PDF
pdf_text = extract_text_from_pdf(pdf_path)
print("Texto extraído do PDF:")
print(pdf_text[:1000])  # Imprimir os primeiros 1000 caracteres para visualização

# Processar o texto com BERT
outputs = process_text_with_bert(pdf_text)
print("Saídas do modelo BERT:")
print(outputs)
```

```python
# Função para criar um grafo com base nas entidades e relações extraídas
def create_graph(entities, relations):
    G = nx.Graph()

    # Adicionar nós ao grafo
    for entity in entities:
        G.add_node(entity)

    # Adicionar arestas ao grafo
    for relation in relations:
        if len(relation) == 2:
            G.add_edge(relation[0], relation[1])

    return G

# Exemplo de entidades e relações extraídas
entities = ["Entity1", "Entity2", "Entity3"]
relations = [("Entity1", "Entity2"), ("Entity2", "Entity3")]

# Criar o grafo
G = create_graph(entities, relations)

# Visualizar o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Disposição dos nós
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=12, font_weight='bold', edge_color='gray')
plt.title("Grafo de Informações")
plt.show()
```

```python
# Instalar bibliotecas necessárias
!pip install transformers pdfplumber networkx matplotlib torch

import networkx as nx
import matplotlib.pyplot as plt
from transformers import BertTokenizer, BertForTokenClassification
from transformers import pipeline
import pdfplumber

# Função para extrair texto de um PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Carregar o modelo e tokenizer para NER
ner_model = BertForTokenClassification.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')
ner_tokenizer = BertTokenizer.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')

# Criar um pipeline de NER
ner_pipeline = pipeline("ner", model=ner_model, tokenizer=ner_tokenizer)

# Função para extrair entidades usando NER
def extract_entities(text):
    entities = ner_pipeline(text)
    entity_dict = {}
    for entity in entities:
        entity_name = entity['word']
        entity_type = entity['entity']
        if entity_name not in entity_dict:
            entity_dict[entity_name] = entity_type
    return entity_dict

# Função para criar um grafo com base nas entidades
def create_graph(entities):
    G = nx.Graph()
    for entity, entity_type in entities.items():
        G.add_node(entity, label=entity_type)

    # Adicionar relações fictícias (você pode adicionar lógica para determinar relações reais)
    for entity1 in entities.keys():
        for entity2 in entities.keys():
            if entity1 != entity2:
                G.add_edge(entity1, entity2)

    return G

# Substitua o caminho do arquivo PDF pelo caminho real
pdf_path = "Art2ao11.pdf"

# Extrair texto do PDF
pdf_text = extract_text_from_pdf(pdf_path)
print("Texto extraído do PDF:")
print(pdf_text[:1000])  # Imprimir os primeiros 1000 caracteres para visualização

# Extrair entidades do texto
entities = extract_entities(pdf_text)
print("Entidades extraídas:")
print(entities)

# Criar o grafo
G = create_graph(entities)

# Visualizar o grafo
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Disposição dos nós
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=12, font_weight='bold', edge_color='gray')
plt.title("Grafo de Entidades Extraídas")
plt.show()
```
