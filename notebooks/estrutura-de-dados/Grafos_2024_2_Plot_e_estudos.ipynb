```python
# Instale as bibliotecas necessárias
!pip install graphviz Pillow

import graphviz
from PIL import Image
import os

# Função para gerar grafo com diferentes estados
def create_graph(state, filename):
    dot = graphviz.Digraph(comment='DFS Example')

    # Definindo os nós
    nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

    # Adicionando nós ao grafo
    for node in nodes:
        if node in state:
            dot.node(node, node, style='filled', fillcolor='lightblue')
        else:
            dot.node(node, node)

    # Adicionando as arestas
    edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('D', 'G')]
    for start, end in edges:
        dot.edge(start, end)

    dot.render(filename, format='png', cleanup=True)

# Caminho para salvar as imagens
filenames = []

# Diferentes estados para o DFS
states = [
    ['A'],             # Frame 1: Visitando A
    ['A', 'B'],       # Frame 2: Visitando B
    ['A', 'B', 'D'],  # Frame 3: Visitando D
    ['A', 'B', 'D', 'G'],  # Frame 4: Visitando G
    ['A', 'B', 'E'],  # Frame 5: Visitando E
    ['A', 'C'],       # Frame 6: Visitando C
    ['A', 'C', 'F']   # Frame 7: Visitando F
]

# Criar imagens
for i, state in enumerate(states):
    filename = f'graph_{i}'
    create_graph(state, filename)
    filenames.append(f'{filename}.png')

# Criar um GIF a partir das imagens
images = [Image.open(fn) for fn in filenames]
gif_path = 'dfs_animation.gif'
images[0].save(gif_path, save_all=True, append_images=images[1:], loop=0, duration=500)

# Mostrar o GIF
from IPython.display import Image as IPImage
IPImage(gif_path)
```

```python
# Instale as bibliotecas necessárias
!pip install graphviz Pillow

import graphviz
from PIL import Image
import os

# Função para gerar grafo com diferentes estados
def create_graph(state, filename):
    dot = graphviz.Digraph(comment='BFS Example')

    # Definindo os nós
    nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

    # Adicionando nós ao grafo
    for node in nodes:
        if node in state:
            dot.node(node, node, style='filled', fillcolor='lightblue')
        else:
            dot.node(node, node)

    # Adicionando as arestas
    edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('D', 'G')]
    for start, end in edges:
        dot.edge(start, end)

    dot.render(filename, format='png', cleanup=True)

# Caminho para salvar as imagens
filenames = []

# Diferentes estados para a BFS
states = [
    ['A'],               # Frame 1: Visitando A
    ['A', 'B', 'C'],    # Frame 2: Visitando B e C
    ['A', 'B', 'C', 'D', 'E'],  # Frame 3: Visitando D e E
    ['A', 'B', 'C', 'D', 'E', 'F'],  # Frame 4: Visitando F
    ['A', 'B', 'C', 'D', 'E', 'F', 'G']  # Frame 5: Visitando G
]

# Criar imagens
for i, state in enumerate(states):
    filename = f'graph_{i}'
    create_graph(state, filename)
    filenames.append(f'{filename}.png')

# Criar um GIF a partir das imagens
images = [Image.open(fn) for fn in filenames]
gif_path = 'bfs_animation.gif'
images[0].save(gif_path, save_all=True, append_images=images[1:], loop=0, duration=500)

# Mostrar o GIF
from IPython.display import Image as IPImage
IPImage(gif_path)
```


A busca em profundidade (DFS) explora um grafo ou árvore priorizando a profundidade antes da largura. Começa em um nó inicial, visitando um filho antes de explorar seus descendentes, recuando ao encontrar um nó sem filhos. É implementada com uma pilha (ou recursão), sendo eficiente em memória, mas podendo ficar presa em caminhos longos ou em ciclos.

A busca em largura (BFS), por outro lado, explora todos os nós em um nível antes de passar para o próximo. Inicia pelo nó raiz e visita todos os vizinhos antes de descer. Utiliza uma fila para manter a ordem de visitação. É ideal para encontrar o caminho mais curto em grafos não ponderados, mas pode ser mais intensiva em termos de memória.
