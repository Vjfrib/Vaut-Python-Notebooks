```python
'''
Tentativa de transliteração

fonte: https://www.editpad.org/tool/br/imagem-em-texto
'''
def maior_soma_quadratico (lista):
    n = len(lista)
    maior_soma = 0

    for i in range(n):
        for j in range(i + 2, n):
            soma = lista[i] + lista[j]
            if soma > maior_soma:
                maior_soma = soma

    return maior_soma
```

```python
'''
dessa vez, com:
(o mesmo SaaS)
'''
def maior_soma_linear(lista):
    n = len(lista)

    if n <= 2:
        return sum(lista)

    dp = [0] * n
    dp[0] = lista[0]
    dp[1] = max(lista[0], lista[1])

    for i in range(2, n):
        dp[i] = max(lista[i] + dp[i-2], dp[i-1])

    return dp[n-1]
```

```python
# ele é bom, se vc identar e corrigir pela img

def maior_soma_nlogn (lista):
    n = len(lista)

    if n <= 2:
        return sum(lista)

    def divide_conquista(lista, inicio, fim):
        if inicio == fim:
            return lista[inicio]
        meio = (inicio + fim) // 2

        esquerda = divide_conquista(lista, inicio, meio)
        direita = divide_conquista(lista, meio+1, fim)

        meio_esquerda = lista[meio]
        soma_esquerda = meio_esquerda

        for i in range(meio-1, inicio-1, -1):
            soma_esquerda += lista[i]
            meio_esquerda = max(meio_esquerda, soma_esquerda)

        meio_direita = lista[meio+1]
        soma_direita = meio_direita

        for i in range(meio+2, fim+1):
            soma_direita += lista[i]
            meio_direita = max(meio_direita, soma_direita)

        return max(meio_esquerda + meio_direita, max (esquerda, direita))

    return divide_conquista (lista, 0, n-1)
```
