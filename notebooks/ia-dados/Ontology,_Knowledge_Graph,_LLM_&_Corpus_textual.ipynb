```python

# Importa as bibliotecas
from pyvis.network import Network
import fitz  # PyMuPDF para leitura de PDFs
from IPython.display import IFrame

# Função para ler o texto do PDF
def read_pdf(file_path):
    text = ""
    pdf_document = fitz.open(file_path)
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    pdf_document.close()
    return text

# Ler o texto do PDF
corpus = read_pdf("L4737compilado.pdf")

# Processar o texto e construir o grafo de conhecimento (implementação simplificada)
def extract_entities_and_relations(text):
    lines = text.split('\n')
    entities = set()
    relations = []

    for line in lines:
        parts = line.split(' ')
        if len(parts) >= 2:
            entities.add(parts[0])
            entities.add(parts[1])
            relations.append((parts[0], parts[1]))

    return entities, relations

entities, relations = extract_entities_and_relations(corpus)

# Criar a visualização interativa
net = Network(notebook=True, height='800px', width='100%', bgcolor='#222222', font_color='white')

# Adicionar nós e arestas
for entity in entities:
    net.add_node(entity, label=entity, color='#1f77b4')

for rel in relations:
    net.add_edge(rel[0], rel[1])

# Salvar e exibir o grafo interativo
net.show('graph.html')

# Exibir o arquivo HTML com o grafo
IFrame(src='graph.html', width=800, height=600)
```
