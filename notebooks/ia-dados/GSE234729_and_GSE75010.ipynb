```python
# -----------------------------
# Instalações (execute apenas se necessário)
# -----------------------------
!pip install GEOparse pandas openpyxl requests

# -----------------------------
# Código Python completo
# -----------------------------
import os
import GEOparse
import pandas as pd
import requests

# -----------------------------
# Configurações
# -----------------------------
gse_id = "GSE234729"  # seu GSE
destdir = "./GSE_data"
os.makedirs(destdir, exist_ok=True)

# -----------------------------
# Baixar o GSE
# -----------------------------
print("Baixando GSE...")
gse = GEOparse.get_GEO(geo=gse_id, destdir=destdir, annotate_gpl=False)

# -----------------------------
# Pegar arquivos suplementares (contagem de RNA-seq)
# -----------------------------
gsm_files = {}
for gsm_name, gsm in gse.gsms.items():
    if "supplementary_file" in gsm.metadata:
        gsm_files[gsm_name] = gsm.metadata["supplementary_file"]

print(f"{len(gsm_files)} arquivos de contagem encontrados.")

# -----------------------------
# Ler todos os arquivos e criar DataFrame
# -----------------------------
expr_dict = {}
for gsm_name, urls in gsm_files.items():
    file_url = urls[0]  # normalmente o primeiro é o arquivo de contagem
    local_file = os.path.join(destdir, file_url.split("/")[-1])

    # baixar se não existir
    if not os.path.exists(local_file):
        print(f"Baixando {local_file} ...")
        r = requests.get(file_url)
        with open(local_file, "wb") as f:
            f.write(r.content)

    # ler arquivo, tentando tab e espaço
    try:
        df = pd.read_csv(local_file, sep="\t", index_col=0)
    except:
        df = pd.read_csv(local_file, sep="\s+", index_col=0)

    # pegar apenas a primeira coluna de contagem
    expr_dict[gsm_name] = df.iloc[:,0]

# -----------------------------
# Criar DataFrame final
# -----------------------------
expr_df = pd.DataFrame(expr_dict)

# -----------------------------
# Salvar em Excel
# -----------------------------
excel_file = os.path.join(destdir, f"{gse_id}_expression.xlsx")
expr_df.to_excel(excel_file)
print(f"✅ Excel criado: {excel_file}")
print(expr_df.head())
```
