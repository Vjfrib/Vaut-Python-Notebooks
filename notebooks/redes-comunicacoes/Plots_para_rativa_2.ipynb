```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os parâmetros do sinal
fs = 1000  # Frequência de amostragem em Hz
t = np.arange(0, 1, 1/fs)  # Vetor de tempo de 0 a 1 segundo

# Criando um sinal composto por três senos de diferentes frequências
freq1 = 50  # Frequência em Hz
freq2 = 70  # Frequência em Hz
freq3 = 90  # Frequência em Hz

# Sinal no domínio do tempo
signal = np.sin(2 * np.pi * freq1 * t) + np.sin(2 * np.pi * freq2 * t) + np.sin(2 * np.pi * freq3 * t)

# Transformada de Fourier para obter o espectro de frequências
fft_signal = np.fft.fft(signal)
fft_freq = np.fft.fftfreq(len(fft_signal), 1/fs)

# Selecionando apenas a parte positiva do espectro
positive_freq_indices = np.where(fft_freq >= 0)
fft_freq = fft_freq[positive_freq_indices]
fft_signal = np.abs(fft_signal[positive_freq_indices])

# Calculando a largura de banda
bandwidth = freq3 - freq1

# Criando o gráfico
plt.figure(figsize=(12, 6))

# Sinal no domínio do tempo
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Sinal no Domínio do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Espectro de frequências
plt.subplot(2, 1, 2)
plt.plot(fft_freq, fft_signal)
plt.title('Espectro de Frequências do Sinal')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.axvline(freq1, color='r', linestyle='--', label=f'Freq1 = {freq1} Hz')
plt.axvline(freq2, color='g', linestyle='--', label=f'Freq2 = {freq2} Hz')
plt.axvline(freq3, color='b', linestyle='--', label=f'Freq3 = {freq3} Hz')
plt.axvspan(freq1, freq3, color='y', alpha=0.3, label=f'Largura de Banda = {bandwidth} Hz')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os parâmetros da senoide
fs = 1000  # Frequência de amostragem (Hz)
T = 1  # Duração do sinal (segundos)
f = 5  # Frequência da senoide (Hz)
t = np.linspace(0, T, fs * T, endpoint=False)  # Vetor de tempo
y = np.sin(2 * np.pi * f * t)  # Sinal da senoide

# Calcular a FFT da senoide
Y = np.fft.fft(y)
Y_magnitude = np.abs(Y) / len(Y)  # Magnitude da FFT
frequencies = np.fft.fftfreq(len(Y), 1 / fs)  # Frequências da FFT

# Plotar os gráficos
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Gráfico da senoide
axs[0].plot(t, y, 'b-', label='Senoide')
axs[0].set_title('Senoide')
axs[0].set_xlabel('Tempo [s]')
axs[0].set_ylabel('Amplitude')
axs[0].legend()
axs[0].grid()

# Gráfico da FFT
axs[1].stem(frequencies[:len(frequencies)//2], Y_magnitude[:len(Y)//2], 'r', markerfmt=" ", basefmt="-b", label='Magnitude da FFT')
axs[1].set_title('Transformada de Fourier (FFT) da Senoide')
axs[1].set_xlabel('Frequência [Hz]')
axs[1].set_ylabel('Magnitude')
axs[1].set_xlim(0, fs/2)
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define time vector
t = np.linspace(-1, 1, 1000)

# Define a signal that is not band-limited: a Gaussian-modulated sine wave
f0 = 50  # frequency of the sine wave
signal = np.sin(2 * np.pi * f0 * t) * np.exp(-t**2 / 0.01)  # Gaussian-modulated sine wave

# Compute the Fourier transform of the signal
signal_fft = np.fft.fftshift(np.fft.fft(signal))
freqs = np.fft.fftshift(np.fft.fftfreq(len(t), d=t[1] - t[0]))

# Plot the time-domain signal
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title('Time-domain Signal (Non Band-limited)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot the frequency-domain representation of the signal
plt.subplot(1, 2, 2)
plt.plot(freqs, np.abs(signal_fft))
plt.title('Frequency-domain Representation')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
bandwidth = 4000  # largura de banda em Hz
levels = [2, 4, 8, 16, 32, 64, 128, 256]  # diferentes números de níveis de tensão
bps = [2 * bandwidth * np.log2(M) for M in levels]

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(levels, bps, marker='o')
plt.axhline(y=8000, color='r', linestyle='--', label='8000 bps')
plt.axvline(x=2, color='g', linestyle='--', label='2 níveis de tensão')

# Adicionando anotações
plt.annotate('4000 Hz largura de banda\n8000 bps com 2 níveis de tensão',
             xy=(2, 8000), xytext=(5, 15000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Relação entre Número de Níveis de Tensão e Taxa de Transmissão')
plt.xlabel('Número de Níveis de Tensão')
plt.ylabel('Taxa de Transmissão (bps)')
plt.grid(True)
plt.legend()
plt.xticks(levels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo os parâmetros
t = np.linspace(0, 1, 1000)  # 1 segundo dividido em 1000 pontos
frequency = 4000  # Frequência em Hz
period = 1 / frequency  # Período da frequência

# Gerando um sinal que alterna entre 0 e 15 V
signal = 15 * (np.sin(2 * np.pi * frequency * t) > 0)

# Plotando o sinal
plt.figure(figsize=(12, 6))
plt.plot(t, signal)
plt.title('Transmissão de 8 Bits de Informação em uma Largura de Banda de 4000 Hz')
plt.xlabel('Tempo (segundos)')
plt.ylabel('Tensão (V)')
plt.grid(True)

# Destacando os ciclos de frequência para ilustrar 8 bits de informação
for i in range(8):
    plt.axvline(x=i*period, color='r', linestyle='--')

plt.show()
```
