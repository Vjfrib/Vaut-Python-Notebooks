```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da Gaussiana
mu = 0       # Média
sigma = 1    # Desvio padrão
t = np.linspace(-5, 5, 1000)  # Intervalo de tempo

# Função Gaussiana
g_t = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((t - mu) / sigma) ** 2)

# Plotar a função Gaussiana
plt.figure(figsize=(10, 4))
plt.plot(t, g_t, label='Função Gaussiana $g(t)$')
plt.title('Função Gaussiana $g(t)$')
plt.xlabel('Tempo $t$')
plt.ylabel('$g(t)$')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.fft import fft, fftfreq

# Calcular a transformada de Fourier
G_f = fft(g_t)
frequencias = fftfreq(len(t), t[1] - t[0])

# Plotar a transformada de Fourier
plt.figure(figsize=(10, 4))
plt.plot(frequencias, np.abs(G_f), label='Transformada de Fourier $G(f)$')
plt.title('Transformada de Fourier da Função Gaussiana $G(f)$')
plt.xlabel('Frequência $f$')
plt.ylabel('$|G(f)|$')
plt.xlim(-10, 10)  # Limitar o eixo x para visualização
plt.legend()
plt.grid()
plt.show()
```

```python
# Parâmetros de amostragem
f_s = 10  # Frequência de amostragem
t_s = np.arange(min(t), max(t), 1/f_s)  # Amostras

# Amostras da função gaussiana
g_t_amostrado = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((t_s - mu) / sigma) ** 2)

# Plotar a função gaussiana amostrada
plt.figure(figsize=(10, 4))
plt.stem(t_s, g_t_amostrado, basefmt=" ", use_line_collection=True, label='Amostras de $g(t)$')
plt.title('Amostragem da Função Gaussiana $g(t)$')
plt.xlabel('Tempo $t$')
plt.ylabel('$g(t)$ amostrado')
plt.grid()
plt.legend()
plt.show()
```

```python
# Calcular a transformada de Fourier da amostragem
G_f_amostrado = fft(g_t_amostrado)
frequencias_amostradas = fftfreq(len(t_s), t_s[1] - t_s[0])

# Plotar o espectro da função amostrada
plt.figure(figsize=(10, 4))
plt.plot(frequencias_amostradas, np.abs(G_f_amostrado), label='Transformada de Fourier da Amostra $G(f)$')
plt.title('Transformada de Fourier da Função Gaussiana Amostrada $G(f)$')
plt.xlabel('Frequência $f$')
plt.ylabel('$|G(f)|$')
plt.xlim(-10, 10)  # Limitar o eixo x para visualização
plt.legend()
plt.grid()
plt.show()
```

```python
# Amostragem no domínio da frequência com impulsos
plt.figure(figsize=(10, 4))
plt.stem(frequencias_amostradas, np.abs(G_f_amostrado), basefmt=" ", use_line_collection=True, label='Espectro Amostrado $G(f)$')
plt.title('Espectro da Função Gaussiana Amostrada $G(f)$ com Impulsos')
plt.xlabel('Frequência $f$')
plt.ylabel('$|G(f)|$ amostrado')
plt.grid()
plt.legend()
plt.show()
```
