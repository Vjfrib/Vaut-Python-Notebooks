````python
import numpy as np
import matplotlib.pyplot as plt

def ustep(t):
    return np.where(t >= 0, 1, 0)

def rect(t):
    return np.where(np.logical_and(np.abs(t) < 0.5, np.abs(t) > -0.5), 1, 0)

def triangl(t):
    return np.where(np.logical_and(np.abs(t) < 1, np.abs(t) > -1), 1 - np.abs(t), 0)

# Teste das funções
t = np.linspace(-2, 2, 1000)

y1 = ustep(t)
y2 = rect(t)
y3 = triangl(t)

# Plotando os gráficos
plt.figure(figsize=(10, 6))

plt.subplot(311)
plt.plot(t, y1)
plt.title('Função degrau unitário')

plt.subplot(312)
plt.plot(t, y2)
plt.title('Função retangular')

plt.subplot(313)
plt.plot(t, y3)
plt.title('Função triangular')

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função degrau unitário
def ustep(t):
    return (t >= 0).astype(int)

# Parâmetros
Dt = 0.002  # Intervalo de tempo (para amostra do sinal)
T = 6       # Período
N = 3       # Para gerar 2M períodos do sinal

# Calculando valores do sinal em um período T
t = np.arange(0, T, Dt)
y = np.exp(-np.abs(t) / 2) * np.sin(2 * np.pi * t) * (ustep(t) - ustep(t - 4))

# Gerando múltiplos períodos
time = []
y_periodic = []

M = 3  # valor arbitrário para M
for i in range(-M, M):
    time.extend(i * T + t)
    y_periodic.extend(y)

# Traçando o gráfico
plt.figure(2)
plt.plot(time, y_periodic, linewidth=2)
plt.xlabel("t")
plt.title("Gráfico de y(t) no domínio do tempo (múltiplos períodos)")
plt.show()

# Cálculo da potência média
y_power = np.sum(np.real(np.conj(y_periodic) * y_periodic)) * Dt / (max(time) - min(time))

# Cálculo da energia em um período T
y_energyT = np.sum(y * np.conj(y)) * Dt
````
